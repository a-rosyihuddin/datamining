# Statistika Deskriptif

##1. Membaca data dari url

```python
import pandas as pd
url = "https://vincentarelbundock.github.io/Rdatasets/csv/carData/WVS.csv"
dataset = pd.read_csv(url)
```

```python
dataset.info()
```

```python
dataset['age']
```

##2. Membuat statistik dari data setiap atribut yang bertype numerik

###2.1 Median

```python
dataset['age'].median()
```

###2.2 Mean

```python
dataset['age'].mean()
```

###2.3 Modus

```python
dataset['age'].mode()
```

##3. Mendeskripsikan kemiringan / skwened dari data setiap variabel

```python
dataset['age'].skew()
```
