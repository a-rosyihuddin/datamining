```python
from scipy.io import arff
import pandas as pd
data = arff.loadarff('/content/drive/MyDrive/Kuliah /Semester 5/datamining/dataset/messidor_features.arff')
df = pd.DataFrame(data[0])
```

```python
df.head()
```

```python
col_names = []
for i in range(20):
    if i == 0:
        col_names.append('quality')
    if i == 1:
        col_names.append('prescreen')
    if i >= 2 and i <= 7:
        col_names.append('ma' + str(i))
    if i >= 8 and i <= 15:
        col_names.append('exudate' + str(i))
    if i == 16:
        col_names.append('euDist')
    if i == 17:
        col_names.append('diameter')
    if i == 18:
        col_names.append('amfm_class')
    if i == 19:
        col_names.append('label')
```

```python
df.columns = [col_names]
df
```

```python
X = df.drop(columns=['label'])
X
```

## Encoder Lebel

```python
y = df['label'].values
y[0:5]
```

```python
from sklearn import preprocessing
le = preprocessing.LabelEncoder()
ybaru = le.fit_transform(y)
ybaru
```

## Splitting Data

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test=train_test_split(X, ybaru, test_size=0.3, random_state=0)
```

## Modeling Dengan KNN Tanpa Normalisasi

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn import metrics
#Membuat k 1 sampai 25
k_range = range(1,26)
scores = {}
scores_list = []
for k in k_range:
        knn = KNeighborsClassifier(n_neighbors=k)
        knn.fit(X_train,y_train)
        y_pred=knn.predict(X_test)
        scores[k] = metrics.accuracy_score(y_test,y_pred)
        scores_list.append(metrics.accuracy_score(y_test,y_pred))
```

```python
scores
```

```python
import matplotlib.pyplot as plt

plt.plot(k_range,scores_list)
plt.xlabel('Value of K for KNN')
plt.ylabel('Testing Accuracy')
```

## Normalisasi Data
### Min-Max Scaler

$$ x' = \frac{x-x_{min}}{x_{max}-x_{min}}$$

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
scaled = scaler.fit_transform(X)
features_names = X.columns.copy()
scaled_features = pd.DataFrame(scaled, columns=features_names)
scaled_features.head()
```

### Save Scaled Data

```python
import joblib
scaler_filename = "scaled.save"
joblib.dump(scaler, scaler_filename) 
```

### Save Data Hasil Normalisasi

```python
import joblib
filename = "normalisasi.sav"
joblib.dump(scaler, filename) 
```

### Spliting Data Setelah Normalisasi

```python
from sklearn.model_selection import train_test_split
X_train_norm, X_test_norm, y_train_norm, y_test_norm=train_test_split(scaled_features, ybaru, test_size=0.2, random_state=1) 
```

## Modeling Data Dengan KNN

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn import metrics
#Membuat k 1 sampai 25
k_range = range(1,26)
scores = {}
scores_list = []
for k in k_range:
        knn = KNeighborsClassifier(n_neighbors=k)
        knn.fit(X_train_norm,y_train_norm)
        y_pred_norm=knn.predict(X_test_norm)
        scores[k] = metrics.accuracy_score(y_test_norm,y_pred_norm)
        scores_list.append(metrics.accuracy_score(y_test_norm,y_pred_norm))
```

```python
scores
```

```python
import matplotlib.pyplot as plt

plt.plot(k_range,scores_list)
plt.xlabel('Value of K for KNN')
plt.ylabel('Testing Accuracy')
```
