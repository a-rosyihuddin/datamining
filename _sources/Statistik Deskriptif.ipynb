# Statistik Deskriptif

```python
import pandas as pd
url = "https://vincentarelbundock.github.io/Rdatasets/csv/carData/WVS.csv"
dataset = pd.read_csv(url)
```

```python
dataset.info()
```

```python
dataset.head()
```

```python
dataset['Equal-Widht'] = pd.cut(dataset['age'], [0, 11, 20, 30, 101], labels = ['Anak', 'Remaja','Dewasa', 'Tua'])
dataset.head(100)
```

```python
dataset['Equal-Frequency'] = pd.qcut(dataset['age'], q=4, precision=1, labels = ['Anak', 'Remaja','Dewasa', 'Tua'] )
dataset.head(100)
```
